## WHO Regionwise Country_name parser

```python
regions =[];Americas =[];Eastern_mediterranean =[];African_region =[];South_east_asia =[];Western_pacific_ =[];Europe =[]
URL = 'countries_data.json'
data_ = pd.read_json(URL)
for i in data_.value:
    if i["WHORegion"]:
        regions.append(i["WHORegion"]["Title"])
        
regions_ = list(set(regions))
print(regions_)

def countries_data(data_):
    for i in data_.value:
        if i["WHORegion"]:
            for j in regions_:
                if i["WHORegion"]["Title"] == j:
                    #print(i["Title"])
                  
                    #print("index : ", regions.index(j))
                    if regions_.index(j) == 0:
                       
                        Americas.append(i["Title"])
                    if regions_.index(j) == 1:
                        Eastern_mediterranean.append(i["Title"])
                    if regions_.index(j) == 2:
                        African_region.append(i["Title"])
                    if regions_.index(j) == 3:
                        South_east_asia.append(i["Title"])
                    if regions_.index(j) == 4:
                        Western_pacific_.append(i["Title"])
                    if regions_.index(j) == 5:
                        Europe.append(i["Title"])
                
                    
countries_data(data_)                
URL2 = 'countries_data2.json'
data_2 = pd.read_json(URL2)
countries_data(data_2) 

URL3 = 'countries_data3.json'
data_3 = pd.read_json(URL3)
countries_data(data_3) 
                      
```

## dataset 1

```python
hiv1["WHO_region"].unique()
```

```python
reg = [];yr = [];deaths = []
who_regions_list = hiv1["WHO_region"].unique()[1:]
def country_names(regions_data, region_values_count):
    
    for i1 in regions_data["WHO_region"].values:
     
        for i in range (region_values_count):
            reg.append(i1)
        
    for i2 in regions_data["Year"].values:
        for i in range (region_values_count):
            yr.append(i2)
            
    for i3 in regions_data["Number of deaths due to HIV/AIDS"].values:
        for i in range (region_values_count):
            deaths.append(i3)
            
for value in who_regions_list:
        
    if value == "Africa":
        region_values_count = len(African_region)
        
    if value == "Americas":
        region_values_count = len(Americas)
        
    if value == "South-East Asia":
        region_values_count = len(South_east_asia)
       
    if value == "Europe":
        region_values_count = len(Europe)
       
    if value == "Eastern Mediterranean":
        region_values_count = len(Eastern_mediterranean)
    
    if value == "Western Pacific":
        region_values_count = len(Western_pacific_)
        
    
        
    regions_data = hiv1[hiv1["WHO_region"] == value]
    country_names(regions_data, region_values_count)  
    
    
```

```python
HIV_dataset1 = pd.DataFrame(list(zip(reg, yr, deaths)), 
               columns =['WHO_region','Year', 'Number of deaths due to HIV/AIDS'])
```

```python
country_column=[]
for i in HIV_dataset1["WHO_region"].unique():
    region_frame = HIV_dataset1[HIV_dataset1["WHO_region"] == i]
    if i == "Africa":
        country_column.append((African_region)*30)
        
    if i == "Americas":
        country_column.append((Americas)*30)
        
    if i == "South-East Asia":
        country_column.append((South_east_asia)*30)
       
    if i == "Europe":
        country_column.append((Europe)*30)
       
    if i == "Eastern Mediterranean":
        country_column.append((Eastern_mediterranean)*30)
    
    if i == "Western Pacific":
        country_column.append((Western_pacific_)*30)
    
```

```python
flatList_countries = [ item for elem in country_column for item in elem]
HIV_dataset1["Countries"] = flatList_countries
```

```python
HIV_dataset1
```

```python
global_data = hiv1[hiv1["WHO_region"] == "Global"]
global_data["Countries"] = "Global"
HIV_map_dataset1 = HIV_dataset1.append(global_data, ignore_index = True)
HIV_map_dataset1.to_csv("HIV1_map_dataset.csv")
```
